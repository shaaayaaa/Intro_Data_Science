```python
from IPython.testing import test
import cv2
import os

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename),cv2.IMREAD_GRAYSCALE)
        if img is not None:
            images.append(img)
    return images
```

```python
X_test=ndarray_test
X_train=ndarray_train

#this lines are for mobilnet
#X_train= np.stack((X_train,)*3, axis=-1)
#X_test= np.stack((X_test,)*3, axis=-1)

print("Training: ", X_train.shape, y_train.shape)
print("Test: ", X_test.shape, y_test.shape)
```

```python
#from keras.utils.np_utils  import to_categorical
from tensorflow.keras.utils  import to_categorical
y_train = to_categorical(y_train, num_classes=None)
y_test = to_categorical(y_test, num_classes=None)
```

```python
num_classes = y_train.shape[1]
print(num_classes)
```

```python
print(y_test.shape)
print(X_test.shape)
print(y_train.shape)
print(X_train.shape)
```

```python

# Custom Model
base_model = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 1)),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(6, activation='softmax')
])

base_model.summary()
```

```python
learning_rate = 0.001
loss_fn = tf.keras.losses.CategoricalCrossentropy(from_logits=False) # from_logits is true when output of model is not probability but a logit(number)
optimizer = keras.optimizers.SGD(learning_rate=learning_rate)
#optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
```

```python
# keras.metrics to get more evaluation metrics
# metric = tf.keras.metrics.TopKCategoricalAccuracy(k=1, name="Top-1 Accuracy")
base_model.compile(loss=loss_fn, optimizer=optimizer, metrics=['accuracy'])
# base_model.compile(loss=loss_fn, optimizer=optimizer, metrics=metric)
```

```python
n = 10 # number of epochs
log = base_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=n, batch_size=256)
```

```python
x = list(range(n))
y = log.history['accuracy']
z = log.history['val_accuracy']

plt.plot(x,y,'r', label = "Training")
plt.plot(x,z,'b', label = "Validation")
plt.legend()
```
